###### Integrated Services Digital Network - ISDN
***

One very popular type of digital wide-area network service is ISDN - stands for Integrated Services Digital Network.

ISDN can be delivered in two different ways: 

i. Basic Rate Interface - BRI


ii. Primary Rate Interface - PRI

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Integrated Services Digital Network - ISDN - Basic Rate Interface - BRI
***

BRI or Basic Rate Interface - may see this referred to as a 2B+D network. 

- ``2B`` stands for two 64 kbit/s bearer channels or ``B`` channels - this is what the data is sent over an ISDN network.

- ``D`` is another channel and this is one 16 kbit/s signally ``D`` channel.

As a signaling channel, the D channel is responsible for setting up the call and breaking down the call and once the call is established, we can send all of our data over one or both of the two bearer channels.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Integrated Services Digital Network - ISDN - Primary Rate Interface - PRI
***

For larger ISDN implementations, would use a PRI connection or Primary Rate Interface. 

This is usually brought into an organization over a T1 or an E1 line. 

- T1 line can support 23 bearer (B) channels and one single (D) channel - ``23B + D". 

- E1 line would support 30 bearer (B) channels and a D channel along with a separate alarm channel - ``30B + D + alarm channel``.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Integrated Services Digital Network - ISDN - Support PSTN PBX Connections inside Organisation
***

Do not usually see ISDN installed as a sole wide-area network connection these days - often brought in to support traditional public switched telephone network PBX connections inside of an organization. 

Might also see this used in radio or broadcasting services where they need to connect two locations very easily and they want to support high-quality audio between those locations.

Some of the more traditional wide-area network connections may be brought in over a T1 or an E1 line. 

< [Table of Contents](#top) | [References](#references) >

***
###### T1 - T-Carrier Level 1
***

T1 stands for T-Carrier Level 1, and it’s a way to connect two locations using time-division multiplexing. 

Commonly see T1 implemented in North America, Japan and South Korea. 

There are 24 channels on a T1 line, and each channel can support 64 kilobits per second for a total of 1.544 megabits per second of a line rate over a T1 connection.

```python
f"T1: {24*64000/1000000} Mbits/s Line Rate."
```

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### E1 - E-Carrier Level 1
***

A similar wide-area connection in Europe is the E1. 

The E1 is slightly different - has 32 channels and each channel can support 64 kilobits per channel.

That gives them a total of a 2.048 megabits per second line rate over E1.

```python
f"E1: {32*64000/1000000} Mbits/s Line Rate."
```

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### T3 - T-Carrier Level 3  - DS3 / E3
***

For these legacy wide-area networks, the next step up from a T1 is a T3 - this stands for T-Carrier Level 3 - might hear this also referred to as a DS3 or Digital Signal 3.

Its usually brought into your facility over coax connections that are connecting to the T3 equipment using these BNC connections.

T3 provides 28 separate T1 circuits for total line rate of 44.736 megabits per second.

```python
f"T3: {28*(24*64000)/1000000} Mbits/s Line Rate."
```

```python
f"T3: {28*24} Channels."
```

In Europe, the E3 combines 16 E1 circuits for total line rate of 34.368 megabits per second. 

```python
f"E3: {16*(32*64000)/1000000} Mbits/s Line Rate."
```

```python
f"E3: {16*32} Channels."
```

Here’s a summary of the T1, the E1, the T3 and the E3 wide-area networks.

The T1 is 24 channels at 64 kbit/s in each channel providing a line rate of 1.544 Mbit/s. 

The E1 is 32 channels, also at 64 kbit/s per second each channel, giving you a total line rate of 2.048 Mbit/s. 

Moving up to the T3 and E3. 

The T3 is 28 T1 circuits, or 672 channels, providing a total line rate of 44.736 Mbit/s. 

The E3 is 16 E1 circuits or 512 total channels for a 34.368 Mbit/s line rate.

###### OC - Synchronous Optical Networking - SONET
***

SONET it is Synchronous Optical Networking and it’s often implemented as rings in a large geographical area. 

There are a number of different line rates on SONET networks. 

These are usually referenced by optical carrier number or OC number.

If a SONET network is an OC-3 network, than it has a line rate of 155.52 Mbit/s. 

OC-12 is 622 Mbit/s - OC-48 moves up to 2.49 Gbit/s - OC-192 SONET network can support a line rate of almost 10 Gbit/s.

###### DSL - Asymmetric Digital Subscriber Line - ADSL
***

A common type of wide-area network connection that we might find in our home is DSL, or what may be more appropriately called ADSL for Asymmetric Digital Subscriber Line. 

This is a wide-area network that allows us to use our existing telephone lines as high-speed digital connections. 

The reason that DSL is asymmetric is because the download speed in DSL is generally much faster than the upload speed.

There’s also a distance limitation associated with DSL - can’t go much farther than about 10,000 feet away from the central office (CO) connection. 

The closer you are to the CO, the higher your line rate will be. 

It’s common to see DSL advertised as having about a 52 Mbit/s downstream and 16 Mbit/s upstream speeds, but those speeds will vary depending on how far away you are from the CO.

###### Metropolitan Area Network - MAN - Metro Ethernet
***

If you need to connect locations within a single city, you might want to consider using metropolitan-area networks such as Metro Ethernet. 

These would be connections that would be in a relatively close geographical area. 

Instead of connecting a T1 line or a T3 line, you can simply have the provider give you an ethernet connection on both ends, and you’re able to connect to your normal ethernet equipment.

Inside of your provider’s network, the transport is probably not going to be ethernet - may be running ethernet over SDH or MPLS or even ethernet over DWDM.

###### Cable Broadband
***

A broadband wide-area network connection is where you are using different frequencies to send data through that connection. 

Could even be using different frequencies for different data types. 

We commonly use broadband when we’re connecting a cable modem to our traditional cable television networks.

This is a standard known as DOCSIS or Data Over Cable Service Interface Specification. 

Depending on your cable providers implementation of DOCSIS, you can see throughput of 4 Mbit/s through 250 Mbit/s and in some areas, they’re supporting gigabit speeds over these cable broadband networks. 

What used to be the traditional cable television company is now also supporting data and voice over the same cable broadband connection.

###### Dialup
***

There are some places where installing a DSL or cable modem connection may not be available. 

You can then use the existing voice telephone lines and connect a modem to be able to send digital signals over these analog voice communication lines. 

These dial-up modems communicate at 56 kbit/s over these voice telephone lines and internally, the modem can compress some data up to 320 kbit/s.

This is obviously a much slower transmission rate than you’d find with the DSL or cable modem connections, and it’s difficult to scale, if you need more connections, you need to bring in more phone lines and more modems. 

Today, we generally see dial-up modems used with very specific use cases.

Maybe you don’t have access to a cable modem or DSL connection or maybe you only need to send a little bit of information only once a day, and you can simply use the existing telephone line to send that data.
